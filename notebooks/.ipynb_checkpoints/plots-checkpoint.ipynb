```python
# %% [markdown]
# # Analyse der Mietpreise: Verschiedene Visualisierungen
# 
# Basierend auf deiner CSV-Datei mit Preisen, Wohnflächen, Bezirken und Ausstattungsgrad.

# %%
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Für schönere Plots
sns.set_style("whitegrid")

# %%
# CSV laden (Pfad anpassen!)
df = pd.read_csv("immowelt_wohnungen_complete.csv")

# %%
# Preis und Wohnfläche in numerisch umwandeln (Strings säubern)
df["Preis_num"] = pd.to_numeric(df["Preis"].str.replace(r"[^\d]", "", regex=True), errors='coerce')
df["Wohnfläche_num"] = pd.to_numeric(df["Wohnfläche"].str.extract(r"(\d+\.?\d*)")[0], errors='coerce')
df["Zimmer_num"] = pd.to_numeric(df["Zimmer"].str.extract(r"(\d+\.?\d*)")[0], errors='coerce')

# Ausstattungsgrad in Zahl (falls nicht schon numerisch)
df["Ausstattungsgrad_num"] = pd.to_numeric(df["Ausstattungsgrad"], errors='coerce')

# %%
# 1. Scatterplot: Preis vs. Wohnfläche, Farbe = Ausstattungsgrad
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Wohnfläche_num", y="Preis_num", hue="Ausstattungsgrad_num", palette="viridis", alpha=0.7)
plt.title("Preis vs. Wohnfläche (Farbe = Ausstattungsgrad)")
plt.xlabel("Wohnfläche (qm)")
plt.ylabel("Preis (€)")
plt.legend(title="Ausstattungsgrad")
plt.show()

# %%
# 2. Hexbin Plot: Dichte von Preis und Wohnfläche
plt.figure(figsize=(10,6))
plt.hexbin(df['Wohnfläche_num'], df['Preis_num'], gridsize=40, cmap='Blues', mincnt=1)
plt.colorbar(label='Anzahl der Wohnungen')
plt.title('Hexbin Plot: Preis vs. Wohnfläche')
plt.xlabel('Wohnfläche (qm)')
plt.ylabel('Preis (€)')
plt.show()

# %%
# 3. Boxplot: Preisverteilung pro Bezirk
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Bezirk', y='Preis_num')
plt.xticks(rotation=45, ha='right')
plt.title('Preisverteilung pro Bezirk')
plt.tight_layout()
plt.show()

# %%
# 4. Interaktiver Scatterplot mit Plotly (nur in Jupyter mit Internetzugang)
fig = px.scatter(df, x='Wohnfläche_num', y='Preis_num', color='Ausstattungsgrad_num',
                 hover_data=['Bezirk', 'Zimmer_num'],
                 title="Interaktiver Scatterplot: Preis vs. Wohnfläche")
fig.show()

# %%
# 5. Scatter Matrix (Pairplot) für wichtige Features
sns.pairplot(df[['Preis_num', 'Wohnfläche_num', 'Zimmer_num', 'Ausstattungsgrad_num']].dropna())
plt.suptitle("Scatter Matrix wichtiger Features", y=1.02)
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("immowelt_wohnungen_complete.csv")

# Preis in Zahl umwandeln (ohne zu speichern)
preis_num = pd.to_numeric(df["Preis"].str.replace(r"[^\d]", "", regex=True), errors="coerce")

# Wohnfläche in Zahl umwandeln (ohne zu speichern)
wohnflaeche_num = pd.to_numeric(df["Wohnfläche"].str.extract(r"(\d+\.?\d*)")[0], errors="coerce")

# Zeilen mit fehlendem Preis oder Wohnfläche finden
fehlerhafte_zeilen = df[preis_num.isna() | wohnflaeche_num.isna()]

# Zeilen mit "Unbekannt" im Bezirk finden
fehlerhafte_zeilen = pd.concat([fehlerhafte_zeilen, df[df["Bezirk"] == "Unbekannt"]]).drop_duplicates()

print(f"Anzahl fehlerhafter Zeilen: {len(fehlerhafte_zeilen)}")
print(fehlerhafte_zeilen)

# Optional: die Indizes der fehlerhaften Zeilen ausgeben
print("Indizes fehlerhafter Zeilen:", fehlerhafte_zeilen.index.tolist())
```
